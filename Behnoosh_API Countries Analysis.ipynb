```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import os
```

```python
# create folder for all outputs

output_folder = 'Outputs'
```

```python
file_path = 'CSV Resources/completed_combined_df.csv'

global_df = pd.read_csv(file_path)
global_df
```

```python
global_df["GDP"] = global_df["GDP"].str.replace(',', '').astype(float)
```

```python
global_df.dtypes
```

```python
# Get the country with the highest number of Masters degrees
max_masters_country = global_df.loc[global_df['2020 Masters or Equivalent Achieved'].idxmax()]['Country']

# Get the country with the highest number of Doctorate degrees
max_doctorate_country = global_df.loc[global_df['2020 Doctorate or Equivalent Achieved'].idxmax()]['Country']

# Get the country with the lowest number of Masters degrees
min_masters_country = global_df.loc[global_df['2020 Masters or Equivalent Achieved'].idxmin()]['Country']

# Get the country with the lowest number of Doctorate degrees
min_doctorate_country = global_df.loc[global_df['2020 Doctorate or Equivalent Achieved'].idxmin()]['Country']

print(f"Country with the highest number of Masters degrees: {max_masters_country}")
print(f"Country with the highest number of Doctorate degrees: {max_doctorate_country}")
print(f"Country with the lowest number of Masters degrees: {min_masters_country}")
print(f"Country with the lowest number of Doctorate degrees: {min_doctorate_country}")
```

```python
# Sort the DataFrame by the number of Masters degrees achieved
global_df_sorted = global_df.sort_values(by='2020 Masters or Equivalent Achieved', ascending=False)

# Plot a bar chart
plt.figure(figsize=(10, 6))
plt.bar(global_df_sorted['Country'], global_df_sorted['2020 Masters or Equivalent Achieved'], label='Masters')
plt.bar(global_df_sorted['Country'], global_df_sorted['2020 Doctorate or Equivalent Achieved'], label='Doctorate', alpha=0.7)

plt.xlabel('Country')
plt.ylabel('Number of Degrees Achieved')
plt.title('Masters and Doctorate Degrees Achieved by Country (2020)')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better visibility
plt.legend()

# Save the figure in the "behnoosh_outputs" folder
plt.savefig(os.path.join(output_folder, 'Masters and Doctorate Degrees Achieved by Country (2020).png'), format='png', bbox_inches='tight')

plt.show()
```

```python
# Sort the DataFrame by GDP
global_df_sorted = global_df.sort_values(by='GDP', ascending=False)

# Plot a grouped bar chart
bar_width = 0.4
index = np.arange(len(global_df_sorted))

plt.figure(figsize=(12, 8))
plt.bar(index, global_df_sorted['2020 Masters or Equivalent Achieved'], width=bar_width, label='Masters')
plt.bar(index + bar_width, global_df_sorted['2020 Doctorate or Equivalent Achieved'], width=bar_width, label='Doctorate')

plt.xlabel('Country')
plt.ylabel('Number of Degrees Achieved')
plt.title('Comparison of Degrees Achieved based on GDP (2020)')
plt.xticks(index + bar_width / 2, global_df_sorted['Country'], rotation=45, ha='right')
plt.legend()

# Save the figure in the "behnoosh_outputs" folder
plt.savefig(os.path.join(output_folder, 'Comparison of Degrees Achieved based on GDP (2020).png'), format='png', bbox_inches='tight')

plt.show()
```

```python
# Sort the DataFrame by 'Population'
global_df_sorted = global_df.sort_values(by='Population', ascending=False)

# Plot a grouped bar chart
bar_width = 0.4
index = np.arange(len(global_df_sorted))

plt.figure(figsize=(12, 8))
plt.bar(index, global_df_sorted['2020 Masters or Equivalent Achieved'], width=bar_width, label='Masters')
plt.bar(index + bar_width, global_df_sorted['2020 Doctorate or Equivalent Achieved'], width=bar_width, label='Doctorate')

plt.xlabel('Country')
plt.ylabel('Number of Degrees Achieved')
plt.title('Comparison of Degrees Achieved based on Population (2020)')
plt.xticks(index + bar_width / 2, global_df_sorted['Country'], rotation=45, ha='right')
plt.legend()

# Save the figure in the "behnoosh_outputs" folder
plt.savefig(os.path.join(output_folder, 'Comparison of Degrees Achieved based on Population (2020).png'), format='png', bbox_inches='tight')

plt.show()
```

1. Masters Degrees vs. Population with Linear Regression:

```python
# Sort the DataFrame by Population
global_df_sorted = global_df.sort_values(by='Population', ascending=False)

# Scatter plot for Masters Degrees vs. Population
plt.figure(figsize=(12, 8))
plt.scatter(global_df_sorted['Population'], global_df_sorted['2020 Masters or Equivalent Achieved'], label='Masters', color='blue', alpha=0.7)

# Linear regression line
slope, intercept, r_value, p_value, std_err = stats.linregress(np.log(global_df_sorted['Population']), np.log(global_df_sorted['2020 Masters or Equivalent Achieved']))
plt.plot(global_df_sorted['Population'], np.exp(intercept + slope * np.log(global_df_sorted['Population'])), color='red', linewidth=2, label='Linear Regression')

# Log scale for better visualization
plt.xscale('log')
plt.yscale('log')

plt.xlabel('Population (log scale)')
plt.ylabel('Number of Master\'s Degrees (log scale)')
plt.title('Masters Degrees vs. Population with Linear Regression')
plt.legend()

# Save the figure in the "behnoosh_outputs" folder
plt.savefig(os.path.join(output_folder, 'Masters Degrees vs. Population with Linear Regression.png'), format='png', bbox_inches='tight')

plt.show()

# Correlation coefficient
correlation_master = r_value
print(f"Correlation between Population and Master's Degrees: {correlation_master:.2f}")
```
