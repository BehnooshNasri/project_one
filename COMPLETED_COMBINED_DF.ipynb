```python
#importing dependencies
import pandas as pd
import requests
from pprint import pprint
```

```python
world_education_df = pd.read_csv("../Documents/output_file.csv")
world_education_df.head()
```

```python
countries_info_df = pd.read_csv("../downloads/countries.csv")
countries_info_df.head()
```

```python
merged_df = pd.merge(world_education_df, countries_info_df, how='inner', on='Country')
merged_df
```

```python
merged_df.replace("..", pd.NA, inplace = True)
merged_df_cleaned = merged_df.dropna()
merged_df_cleaned
```

```python
merged_df_cleaned = merged_df_cleaned.drop(columns = "Country Code_x")
merged_df_cleaned
```

```python
merged_df_cleaned = merged_df_cleaned.drop(columns = "Country Code_y")
merged_df_cleaned
```

```python
merged_df_cleaned = merged_df_cleaned.rename(columns={'Description_x': 'Masters or Equivalent Achieved',
                                                      'Description_y Name_y': 'Doctorate or Equivalent Achieved'})
merged_df_cleaned
```

```python
columns_to_drop = ['Masters or Equivalent Achieved', 'Doctorate or Equivalent Achieved']
merged_df_cleaned = merged_df_cleaned.drop(columns=columns_to_drop)
merged_df_cleaned
```

```python
merged_df_cleaned.columns
```

```python
merged_df_cleaned = merged_df_cleaned.rename(columns={'2020': '2020 Doctorate or Equivalent Achieved'})
merged_df_cleaned
```

```python
merged_df_cleaned.to_csv('completed_combined_df.csv', index=False)
```
