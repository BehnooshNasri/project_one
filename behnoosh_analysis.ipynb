```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = 'Gender Type and Education Level.csv'

initial_df = pd.read_csv(file_path)
initial_df
```

```python
print(initial_df.iloc[:, 0].unique())
```

```python
# Get unique values from the first column
unique_values = initial_df["Province/Territory"].unique()

# Create a dictionary to store DataFrames
dfs_by_province = {}

# Iterate through unique values and create DataFrames
for value in unique_values:
    # Filter the original DataFrame for the current value
    df_province = initial_df[initial_df["Province/Territory"] == value]
    
    # Store the DataFrame in the dictionary
    dfs_by_province[value] = df_province

# dfs_by_province['Newfoundland and Labrador']
dfs_by_province['Prince Edward Island']
# dfs_by_province['Nova Scotia']
# dfs_by_province['New Brunswick']
# dfs_by_province['Quebec']
# dfs_by_province['Ontario']
# dfs_by_province['Manitoba']
# dfs_by_province['Alberta']
# dfs_by_province['British Columbia']
# dfs_by_province['Yukon']
# dfs_by_province['Northwest Territories']
# dfs_by_province['Nunavut']
```

```python
print(dfs_by_province['Prince Edward Island']['Description'].unique())
```

```python
pei = dfs_by_province['Prince Edward Island'][["Description", "Total", "Male", "Female", "% of Provincial Population", "Male %", "Female %"]]
pei.set_index("Description", inplace=True)
pei
```

```python
# #if it's better to flip rows and columns
#pei_transposed = pei.transpose()
# pei_transposed
```

```python
pei.describe()
```

```python
# Bar plot for Total, Male, and Female
plt.figure(figsize=(12, 6))
pei[['Total', 'Male', 'Female']].plot(kind='bar', stacked=True)
plt.title('Population Distribution by Gender')
plt.xlabel('Education Level')
plt.ylabel('Population')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Line plot for % of Provincial Population, Male %, and Female %
plt.figure(figsize=(12, 6))
pei[['% of Provincial Population', 'Male %', 'Female %']].plot(kind='line', marker='o')
plt.title('Percentage Distribution')
plt.xlabel('Education Level')
plt.ylabel('Percentage')
plt.xticks(rotation=45, ha='right')
plt.show()
```
