```python
#importing dependencies
import pandas as pd
import requests
from pprint import pprint
```

```python
world_master_1_df = pd.read_csv("../documents/output_file.csv")
world_master_1_df.head()
```

```python
world_master_1_df = world_master_1_df.drop(columns = "Series Code")
world_master_1_df
```

```python
world_doctorate_df = pd.read_csv("../downloads/Doctorate_enrollment.csv")
world_doctorate_df.head()
```

```python
world_doctorate_df = world_doctorate_df.drop(columns = "Series Code")
world_doctorate_df
```

```python
world_doctorate_df = world_doctorate_df.drop(columns = "Country Code")
world_doctorate_df
```

```python
merged_df = pd.merge(world_master_1_df, world_doctorate_df, how='inner', on='Country Name')
merged_df
```

```python
columns_to_drop = ['Country Code', 'Country Name']
merged_df = merged_df.drop(columns=columns_to_drop)
merged_df
```

```python
merged_df = merged_df.rename(columns={'2020 Masters': 'Masters or Equivalent Achieved',
                                      '2020 Doctorate or Equivalent Achieved': 'Doctorate or Equivalent Achieved'})
merged_df
```
