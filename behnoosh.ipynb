```python
#importing dependencies
import pandas as pd
import requests
from pprint import pprint
```

```python
base_url = "http://api.worldbank.org/v2/country"
endpoint = "?format=json"
combined_url = f"{base_url}{endpoint}"
params = {"per_page": 50, "page": 1}  

# Initialize a list to store country names
countries_list = []

# Initial request to get the total number of pages
response = requests.get(combined_url, params=params)
total_pages = response.json()[0]["pages"]

# Fetch data from all pages
for page in range(1, total_pages + 1):
    params["page"] = page
    response = requests.get(combined_url, params=params)
    
    if response.status_code == 200:
        data = response.json()
        
        for country_data in data[1]:
            country_name = country_data["name"]
            
            # Append country name to the list
            countries_list.append(country_name)
    else:
        print(f"Error on page {page}: {response.status_code}")

```

```python
# Create a DataFrame
world_countries = pd.DataFrame({"Countries": countries_list})
world_countries.sample(15)
```
