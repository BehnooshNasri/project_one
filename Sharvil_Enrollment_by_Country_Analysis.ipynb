```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv = 'Source Resources/census_enrollment_countries.csv'

main_df = pd.read_csv(csv)
main_df
```

```python
main_df.dropna(inplace=True)
main_df.sample(5)
```

```python
canada_df = main_df[['Geography','Canada','Canada.1','Canada.2','Canada.3','Canada.4','Canada.5']]
canada_df.columns = canada_df.iloc[0].tolist()
canada_df = canada_df[1:]
canada_df.replace('..','0',inplace=True)
canada_df.reset_index(drop=True,inplace=True)
canada_df
```

```python
canada_df_columns = canada_df.columns.tolist()
canada_df_column_names ={
    0 : 'Country of citizenship',
    1: "All Post-Secondary 2020/2021" ,
    2: "All Post-Secondary 2021/2022",
    3: "Master's or equivalent 2020/2021",
    4: "Master's or equivalent 2021/2022",
    5: "Doctoral or equivalent 2020/2021",
    6: "Doctoral or equivalent 2021/2022"}

for index,new_name in canada_df_column_names.items():
    canada_df_columns[index] = new_name
    
canada_df.columns = canada_df_columns
canada_df.set_index(canada_df.columns[0],inplace=True)
canada_df
```

```python
numeric_columns = canada_df.columns[1:]  # Assuming the first column is non-numeric
for column in numeric_columns:
    canada_df[column] = canada_df[column].str.replace(',', '').astype(int)
```

```python
canada_sorted_df_2020_2021_masters = canada_df.sort_values(by="Master's or equivalent 2020/2021", ascending = False)
canada_sorted_df_2020_2021_masters = canada_sorted_df_2020_2021_masters.drop(canada_sorted_df_2020_2021_masters.index[:2])
canada_sorted_df_2020_2021_masters = canada_sorted_df_2020_2021_masters.sort_values(by="Master's or equivalent 2020/2021", ascending = False)
canada_sorted_df_2020_2021_masters.reset_index(inplace=True)
canada_sorted_df_2020_2021_masters['Country of citizenship'] = canada_sorted_df_2020_2021_masters['Country of citizenship'].str.split(' \[', expand=True)[0]
canada_sorted_df_2020_2021_masters.set_index('Country of citizenship', inplace=True)
canada_sorted_df_2020_2021_masters= canada_sorted_df_2020_2021_masters.replace(',', '', regex=True).astype(int)
canada_sorted_df_2020_2021_masters.head(10)
```

```python
top_ten_countries = canada_sorted_df_2020_2021_masters.iloc[:10,2]


plt.figure(figsize=(10, 6)) 
plt.bar(top_ten_countries.index, top_ten_countries.values)

plt.xlabel('Countries')  
plt.ylabel('Enrollment Count')  
plt.title("Top 10 Countries with Master's or equivalent Enrollments - Fall 2020/2021")  
plt.xticks(rotation=45,ha='right')  

plt.show() 
```

```python
plt.savefig("Outputs/Master's or equivalent 2020-2021 Enrollments for Top 10 Countries.png")
```
