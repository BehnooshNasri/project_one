```python
from pathlib import Path
import pandas as pd
import numpy as np
```

```python
csv = '98-401-X2021001_English_CSV_data.csv'

main_df = pd.read_csv(csv,encoding='latin-1')
main_df
```

```python
# Gathering only provincial data
filtered_by_prov = main_df[(main_df["GEO_LEVEL"] == 'Territory') | (main_df["GEO_LEVEL"]=="Province")]
filtered_by_prov = filtered_by_prov[['GEO_LEVEL','GEO_NAME','CHARACTERISTIC_NAME','CHARACTERISTIC_ID','C1_COUNT_TOTAL','C2_COUNT_MEN+','C3_COUNT_WOMEN+','C10_RATE_TOTAL','C11_RATE_MEN+','C12_RATE_WOMEN+']]
filtered_by_prov
```

```python
population_df = filtered_by_prov[filtered_by_prov['CHARACTERISTIC_ID'] == 1]
population_df
```

```python
# Drop NaN 
cleaned_df = filtered_by_prov.dropna()
cleaned_df
```

```python
# Grabbing the characteristic IDs we want
education_df = cleaned_df[(cleaned_df['CHARACTERISTIC_ID'] == 8) |
                        (cleaned_df['CHARACTERISTIC_ID'] == 1998) | 
                        (cleaned_df['CHARACTERISTIC_ID'] == 2012) | 
                        (cleaned_df['CHARACTERISTIC_ID'] == 2013) |
                        (cleaned_df['CHARACTERISTIC_ID'] == 2014) |
                        (cleaned_df['CHARACTERISTIC_ID'] == 2028) |
                        (cleaned_df['CHARACTERISTIC_ID'] == 2029) ]

education_df.loc[education_df['CHARACTERISTIC_ID'] == 8, 'CHARACTERISTIC_NAME'] = 'Total population'

# Rename columns
education_df = education_df.rename(columns ={
    'GEO_NAME' : 'Province/Territory',
    'CHARACTERISTIC_NAME' : 'Description',
    'C1_COUNT_TOTAL':'Total',
    'C2_COUNT_MEN+': 'Male',
    'C3_COUNT_WOMEN+': 'Female',
    'C10_RATE_TOTAL' : '% of Provincial Population',
    'C11_RATE_MEN+' : 'Male %',
    'C12_RATE_WOMEN+' : 'Female %'
                     })
# Resetting the index and pulling only columns we want
education_df.reset_index(inplace=True)
education_df.set_index("Province/Territory",inplace=True)
education_df = education_df[['Description','Total','Male','Female','% of Provincial Population','Male %','Female %']]
education_df
```

```python
# Convert to integer
education_df['Male'] = education_df['Male'].astype(int)
education_df['Female'] = education_df['Female'].astype(int)
education_df['Total'] = education_df['Total'].astype(int)

# Displaying the DF
education_df
```

```python
education_df.to_csv('Gender Type and Education Level.csv')
```
