```python
from pathlib import Path
import pandas as pd
import numpy as np
```

```python
csv = '98-401-X2021001_English_CSV_data.csv'

main_df = pd.read_csv(csv,encoding='latin-1')
main_df
```

```python
filtered_by_prov = main_df[(main_df["GEO_LEVEL"] == 'Territory') | (main_df["GEO_LEVEL"]=="Province")]
filtered_by_prov = filtered_by_prov[['GEO_LEVEL','GEO_NAME','CHARACTERISTIC_NAME','CHARACTERISTIC_ID','C1_COUNT_TOTAL','C2_COUNT_MEN+','C3_COUNT_WOMEN+','C10_RATE_TOTAL','C11_RATE_MEN+','C12_RATE_WOMEN+']]
filtered_by_prov
```

```python
cleaned_df = filtered_by_prov.dropna()
cleaned_df = cleaned_df.drop_duplicates()
cleaned_df
```

```python
# removing spaces before the words in Characteristic Name column
cleaned_df.loc[:,'CHARACTERISTIC_NAME'] = cleaned_df['CHARACTERISTIC_NAME'].str.strip()
```

```python
education_df = cleaned_df[(cleaned_df['CHARACTERISTIC_NAME'] == 'Earned doctorate') | 
                         (cleaned_df['CHARACTERISTIC_NAME'] == "Master's degree")]
education_df
```
